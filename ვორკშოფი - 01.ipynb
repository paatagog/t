# კომპიუტერული მეცნიერების საგანმანათლებლო კვირეული

ვორკშოფი - ხელოვნური ინტელექტი.
დაპროგრამების ენა Python-ზე ვწერთ ხელნაწერი ციფრების ამოცნობის პროგრამას. პროგრამის გეგმა არის ასეთი:

1. გამზადებული მონაცემების ჩამოტვირთვა.
2. მანქანური სწავლების მოდელის შექმნა სურათების კლასიფიკაციისთვის.
3. ნეირონული ქსელის გაწვრთნა.
4. მიღებული მოდელის აკურატულობის შემოწმება.

## TensorFlow-ს გამოძახება

გამოვიყენოთ Import ბრძანება TensorFlow-ს გამოსაძახებლად:

```
import tensorflow as tf
```

## მონაცემთა სიმრავლის ჩატვირთვა

ჩამოვტვირთოთ [MNIST](http://yann.lecun.com/exdb/mnist/) მონაცემთა სიმრავლე და [0, 255] რიცხვების მაგივრად გადავაქციოთ იგი [0, 1] რიცხვებად:

```
mnist = tf.keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

## ავაგოთ მანქანური სწავლების მოდელი

ავაგოთ `tf.keras.Sequential` მოდელი შრეების ერთმენეთზე მიწყობით.

```
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10)
])
```

თითოეული სურათისთვის მოდელი მისი შესაბამისი კლასის ნომერს მოგვცემს.

```
predictions = model(x_train[:1]).numpy()
predictions
```

განვსაზღვროთ დანაკარგის ფუნქცია `losses.SparseCategoricalCrossentropy`

```
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
```

```
model.compile(optimizer='adam',
              loss=loss_fn,
              metrics=['accuracy'])
```

## გავწვრთნათ და შევამოწმოთ ჩვენი მოდელი

`Model.fit`-ის საშუალებით ოპტიმალურ წონებს ვიპოვით და შევამცირებთ დანაკარგის ფუნქციას.

```
model.fit(x_train, y_train, epochs=5)
```

`Model.evaluate` ამოწმებს მოდელის საიმედოობას საცდელ მონაცემებზე

```
model.evaluate(x_test,  y_test, verbose=2)
```

სურათების დახარისხება  დაახლოებით 98% საიმედოობით გამოგვივიდა.

```
model.predict(x_test[1:2])
```
